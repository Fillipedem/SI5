```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d, Axes3D #<-- Note the capitalization! 
from sklearn.datasets import load_iris
from kmeans import Kmeans
```

#### load/run Kmeans no dataset

```python
# Dataset
iris = load_iris()
# Run Kmeans
kmeans = Kmeans(iris.data, 3)
clusters = []
for i in range(15):
    clusters = kmeans.one_step()
```

#### Plotando os pontos da base de dados Iris, removendo umas das dimensÃµes por plot

```python
for i in range(0, 4):
    data = np.delete(iris.data, i, 1)
    centroids = np.delete(kmeans.centroids, i, 1)
    
    # plot first 3 columns
    colors = {0: 'red', 1: 'blue', 2: 'green'}

    fig = plt.figure()
    ax = Axes3D(fig)

    # print points
    for idx, ex in enumerate(data):
        x, y, z = ex
        ax.scatter(x, y, z, color=colors[iris.target[idx]])

    # print centroids
    for c in centroids:
        ax.scatter(c[0], c[1], c[2], color='black')
```

```python
# check classes
print("O quanto cada cluster formado contÃ©m cada a classe do problema Iris:")
print(iris.target_names)
for c in clusters:
    print("Cluster ", c, ":")
    classes = [0, 0, 0]
    for i in clusters[c]:
        classes[iris.target[i]] += 1
    print(classes)
```
